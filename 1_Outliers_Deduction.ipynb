```

import pandas as pd
import numpy as np
from google.colab import files
import io
```

```

uploaded=files.upload()
```

```
df=pd.read_csv(io.BytesIO(uploaded['heights.csv']))
df
```

```
maximum_threshold=df['height'].quantile(0.95)
```

```
df[df['height']>maximum_threshold]
```

```
minium_threshold=df['height'].quantile(0.05)
```

```
df[df['height']<minium_threshold]
```

```
df[(df['height']<maximum_threshold)&(df['height']>minium_threshold)]
```

```
df=pd.read_csv(io.BytesIO(uploaded['bhp.csv']))
df
```

```
df.describe()
```

```
min_threshold, max_threshold = df.price_per_sqft.quantile([0.001,0.999])
min_threshold, max_threshold
```

```
df[df.price_per_sqft<min_threshold]
```

```
df[df.price_per_sqft>max_threshold]
```

```
df2=df[(df.price_per_sqft>min_threshold)&(df.price_per_sqft<max_threshold)]
df2.shape

```
